### Building a visualization for our dataset

```
def plot_prediction(train_data = x_train 
                   ,train_label = y_train 
                   ,test_data = x_test 
                   ,test_label = y_test
                   ,prediction = None):
    
    ''' This is use to ploting the train and test dat'''
    plt.figure(figsize=(10,7))

    # plotting train data in blue
    plt.scatter(x_train , y_train , c = 'b' , s = 4 , label = 'Training Data')

    # plotting test data in red
    plt.scatter(x_test , y_test , c = 'r' , s = 4 , label = 'Testing Data')

    # plotting the prediction
    if prediction is not None:
        plt.scatter(test_data , prediction , c = 'g' ,s = 4, label = 'Predictions')
    
    plt.legend()
```

### Building Model

* Our first pytorch Model
**Resources**
https://pytorch.org/docs/stable/generated/torch.nn.Module.html
https://pytorch.org/docs/stable/generated/torch.nn.parameter.Parameter.html

**Needs to Understand**
* Gradient Descent
* Backpropogation

```
import torch
import torch.nn as nn

# create Linear regression model class
class LinearRegression(nn.Module):
    #-> in pytorch everything is inherited from nn.Module
    def __init__(self):
        super().__init__()  # corrected line
        self.weights = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float))
        self.bias = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float))

    # Now defining the forward method for forward computations
    def forward(self, x: torch.Tensor) -> torch.Tensor:  # -> forward method must be overwritten
        return self.weights * x + self.bias  # where x is the input data
```

### Pytorch Model Building Essential

* `torch.nn` - contains all of the building for computation graph(neural network)
* `torch.nn.Parameter` - what parameter should our model try and learn , often a pytorch layer will set for use
* `torch.nn.Module` - This is the base class for all neural network modules,if you subclass it you should overwrite forward() method
* `torch.optim` - This is where the optimizer live , and help us with gradient descent
* `def forward()` - All nn.Module subclass require you to overwrite forward(), this method define what happen in the forward computation
* `Cheet sheet` - https://pytorch.org/tutorials/beginner/ptcheat.html and https://www.datacamp.com/cheat-sheet/deep-learning-with-py-torch

### 3. Train model

* The whole idea of training is for a model to move from some **unknown** parameters
* Or in other words a poor representation of data to a better representation
* One way to measure how poor or wrong your prediction are is to use loss function
* **Note:** Loss function may also be called cost function or criterion in different area .For our case we're going to refer to it as a loss function

Thing we need to train:
* **Loss:** A fuction measure how wrong your model prediction are to ideal output , lower is better
* **Optimizer:** Take into acount the loss of the model and adjust the model's paramter(weight and bias) to improve the loss function 
->  1. paramas : Takes the parameters of the model that need to be adjust
->  2. lr : learning rate

And specially for pytorch , we need:
* A training loop
* A testing loop

```
# Setting up Loss function
loss_fn = torch.nn.L1Loss()

# setting up optimizer
optimizer = torch.optim.SGD(params = model_o.parameters() , lr = 0.01)
```

<img src="image-20240316-094725.png" width="" align="" />

### Buidling a training and testing loop in pytorch
A couple of things we need in a training loops:

0. Loop through data
1. Forward pass 
2. calculate the loss 
3. Optimize zero grads
4. Loss backward - move backward through network to calcualte the gradient of each of the parameter of our model with respect to loss (**backpropogation**)
5. Optimize step - use the optimizer to adjust our model parameter to try and improve the loss (**gradient descent**)

```
list(model_o.named_parameters())
```

```
torch.manual_seed(42)
# step up epoch
epochs = 100

epoch_counts = []
loss_val = []
test_loss_val = []

# loop through data
for epoch in range(epochs):
    model_o.train() # train model in pytorch set all paramter that require gradients to require gradient

    # forward pass
    y_pred = model_o(x_train)

    # loss calculation
    loss = loss_fn(y_pred , y_train)
    # print("Loss:",loss)

    # optimizer grad
    optimizer.zero_grad()

    # perform backward propogation with respect to model parameter
    loss.backward()

    # perform gradient descent
    optimizer.step()

    # testing code
    # turn off gradient tracking because not need for evaluation and testing
    model_o.eval()
    with torch.inference_mode():
        # Do the forward pass for testing 
        y_prediction = model_o(x_test)

        # calculate the loss basd on y_prediction and y_test
        loss_ = loss_fn(y_prediction,y_test)
    
    # printing all the necessary details
    if epoch % 10 == 0:
        epoch_counts.append(epoch)
        loss_val.append(loss)
        test_loss_val.append(loss_)
        print(f"Epochs : {epoch} | Loss : {loss} | TestLoss : {loss_}")
    # print(model_o.state_dict())
```

```
model_o.state_dict()
```

```
 with torch.inference_mode():
    y_pred_new = model_o(x_test)

plot_prediction(prediction=y_pred_new)
loss
```

```
plt.plot(epoch_counts , list(torch.tensor(loss_val).numpy()) , label = 'Train Loss')
plt.plot(epoch_counts , test_loss_val , label = 'Test Loss')
plt.title("Epoch vs Loss")
plt.ylabel("Loss")
plt.ylabel("Epochs")
plt.legend()
```

### Creating your own Linear Model using what we have done so far

```
import torch
import torch.nn as nn
weight = 1
bias = 1

start = 1
end = 2
step = 0.2
x = torch.arange(start , end , step).unsqueeze(dim = 1)
y = x * weight + bias
```

```
# creating training and testing dataset
train_split = int(0.8*len(x))
x_train = x[:train_split]
y_train = y[:train_split]
x_test = x[train_split:]
y_test = y[train_split]
```

```
class LinearRegressionV2(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_layer = nn.Linear(1,1)
    
    def forward(self,x : torch.Tensor) -> torch.Tensor:
        return self.linear_layer(x)
```

```
torch.manual_seed(42)
model_1 = LinearRegressionV2()
model_1.state_dict()
```

```
# setting up the loss function but this time we will use MSELoss function
loss_fn = torch.nn.MSELoss()

# setting up the optimizer but this time we will use adam optimizer
optimizer = torch.optim.Adam(params = model_1.parameters() , lr = 0.01)
```

```
epochs_list = []
train_loss_list = []
test_loss_list = []
```

```
model_1.state_dict()
```

```
import torch

torch.manual_seed(42)

epochs = 250
epochs_list = []
train_loss_list = []
test_loss_list = []

for epoch in range(epochs):
    model_1.train()
    y_preds = model_1.forward(x_train)
    train_loss = loss_fn(y_preds, y_train)
    optimizer.zero_grad()
    train_loss.backward()
    optimizer.step()

    model_1.eval()
    with torch.no_grad():
        y_pred = model_1.forward(x_test)
        test_loss = loss_fn(y_pred, y_test)

    if epoch % 10 == 0:
        epochs_list.append(epoch)
        train_loss_list.append(train_loss.item())
        test_loss_list.append(test_loss.item())
```

```
train_loss_list , test_loss_list , epochs_list
```

```
import matplotlib.pyplot as plt
def plot_prediction(train_data = x_train 
                   ,train_label = y_train 
                   ,test_data = x_test 
                   ,test_label = y_test
                   ,prediction = None):
    
    ''' This is use to ploting the train and test dat'''
    plt.figure(figsize=(10,7))

    # plotting train data in blue
    plt.scatter(x_train , y_train , c = 'b' , s = 4 , label = 'Training Data')

    # plotting test data in red
    plt.scatter(x_test , y_test , c = 'r' , s = 4 , label = 'Testing Data')

    # plotting the prediction
    if prediction is not None:
        plt.scatter(test_data , prediction , c = 'g' ,s = 4, label = 'Predictions')
    
    plt.legend()
```

```
plot_prediction(train_loss_list,test_loss_list)
```
